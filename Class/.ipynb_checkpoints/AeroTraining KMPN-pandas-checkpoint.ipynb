# Conditional Filtering
Setelah kita tahu cara melakukan indexing dan seleksi data dari row dan column, apakah kita bisa melakukan filtering atau seleksi data dengan kriteria/kondisi yang lebih spesifik?

Tentunya bisa, dengan Conditional Filtering
(Melakukan filtering data dengan kondisi yang kita inginkan)

```python
df = pd.read_csv('tips.csv')
df.head()
```

## Defining a Condition

```python
df['total_bill']<15
```

```python
df[df['total_bill']<15]
```

```python
df[df['smoker']=='Yes']
```

## Defining Multiple Conditions

### And Condition `&`

```python
df[(df['total_bill'] < 15) & (df['smoker']=='Yes')]
```

### Or Condition`|`

```python
df[(df['day'] =='Sun') | (df['day']=='Sat')]
```

### Using Negation `~`

```python
df[(df['total_bill'] < 15) & ~(df['smoker']=='Yes')]
```

### Using not equals to `!=`

```python
df[(df['total_bill'] < 15) & (df['smoker']!='Yes')]
```

### Conditional Operator `.isin()`

```python
my_days = ['Fri','Sat','Sun']
df['day'].isin(my_days)
```

```python
df[df['day'].isin(['Fri','Sat','Sun'])]
```

## `.apply()`
Dapat digunakan untuk meng `apply` suatu fungsi yang kita definisikan ke selected data pada DataFrame

```python
def last_four(num):
    return str(num)[-4:]

print(last_four('123456789'))
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
df.head()
```

```python
def doll_tag(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 20:
        return '$$'
    else:
        return '$$$'
```

```python
df['Expensive'] = df['total_bill'].apply(doll_tag)
df.head()
```

```python
df['total_bill'].apply(lambda bill:bill*2)
```

```python
def quality(total_bill,tip):
    if tip/total_bill  > 0.15:
        return "Generous"
    else:
        return "Other"
```

```python
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
df.head()
```

`np.vectorize` for this is much faster than using apply as above

```python
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
df.head()
```

##`.groupby()`

Kita perlu mendefinisikan metode aggregat seperti apa yang kita inginkan untuk datanya.

Yang umum digunakan adalah

* mean(): Compute mean of groups
* sum(): Compute sum of group values
* size(): Compute group sizes
* count(): Compute count of group
* std(): Standard deviation of groups
* var(): Compute variance of groups
* sem(): Standard error of the mean of groups
* describe(): Generates descriptive statistics
* first(): Compute first of group values
* last(): Compute last of group values
* nth() : Take nth value, or a subset if n is a list
* min(): Compute min of group values
* max(): Compute max of group values

    Documentation: https://pandas.pydata.org/docs/reference/groupby.html

```python
avg_by_year = df.groupby('model_year').mean()
avg_by_year
```

## Cleaning Strings
Data yang di input oleh user pada dunia nyata biasa banyak yang mengandung typo ataupun elemen-elemen yang tidak diinginkan, sehingga perlu dilakukan cleaning atau membersihkan data teks agar bersih dari typo atau hal-hal non esensial yang ingin dibuang dari teks

```python
typo_names = pd.Series(["   1khwan  ","ic;al   ","  Lucky"])
typo_names 
```

```python
typo_names.str.replace(";","")
```

```python
typo_names.str.replace(";","").str.replace("1","i").str.strip()
```

```python
typo_names.str.replace(";","").str.replace("1","i").str.strip().str.capitalize()
```

## Alternative Cleaning with Custom `.apply()`

```python
def cleanup(name):
    name = name.replace(";","")
    name = name.replace("1","i")
    name = name.strip()
    name = name.capitalize()
    return name
```

```python
typo_names
```

```python
typo_names.apply(cleanup)
```
